#### *Exercise 6.6*

#### In Example 6.2 we stated that the true values for the random walk example are 1/6, 2/6, 3/6, 4/6, and 5/6, for states A through E. Describe at least two different ways that these could have been computed. Which would you guess we actually used? Why?

---
#### Resposta 1:

A primeira forma seria usando DP, uma vez que é possível modelar o ambiente e obter as probabilidades conforme uma MDP. A segunda seria calculando matematicamente, uma forma simples que eu encontrei foi a seguinte: tem-se $N$ passos para a recompensa $1$ e $6 - N$ passos para a recompensa zero; uma vez que a probabilidade de transição para direita ou esquerda é a mesma, uma forma simples de calcular é considerando que a probabilidade tem relação com a distância para os estados terminais. Sendo mais preciso pode-se considerar um mesmo fator $p$ para ambos os destinos (recompensa 1 ou 0) multiplicados pelas respectivas distâncias, a distância é conhecida, mas o termo deve ser calculado e junto com a distância informar a probabilidade. Como se tratam de probabilidades: $p/N + p/(6-N) = 1$ e $p/N$ é o valor desejado.

Considerando A (N = 5):

$p/5 + p/1 = 1$

$p = 5/6$, mas $p/N = 1/6.$

Para B (N = 4):

$p/4 + p/2 = 1$

$p = 4/3$, mas $p/N = 1/3 = 2/6$

Para C (N = 3):

$p/3 + p/3 = 1$

$p = 3/2$, mas $p/N = 1/2 = 3/6$

Para D (N = 2):

$p/2 + p/4 = 1$

$p = 4/3$, mas $p/N = 4/6$

Para E (N = 1):

$p + p/5 = 1$

$p = 5/6$ e $p/N = 5/6.$

O livro deve ter optado pelo calcula de probabilidades, já que é mais simples.
