#### *Exercise 5.1*

#### Consider the diagrams on the right in Figure 5.1. Why does the estimated value function jump up for the last two rows in the rear? Why does it drop off for the whole last row on the left? Why are the frontmost values higher in the upper diagrams than in the lower?

---
#### Resposta 1:
i) Segundo o livro, o jogador quando tem 20 ou 21 opta por não pegar mais nenhuma carta, 21 ele ganha e 20 está próximo de ganhar. Qualquer valor abaixo de 20, o jogador solicita mais uma carta. A medida que ele solicita mais cartas ao dealer, tanto pode atingir de fato 21 quanto ultrapassar esse valor, pela figura 5.1 é possível ver que dificilmente ele vai conseguir vencer caso a soma esteja entre 12 e 19. 

ii) Porque caso a carta do dealer virada para cima seja um ace, basta ele tirar 10 na carta virada e na pior hipótese garante um empate. Caso a carta para cima seja 10, basta que a carta virada para baixo seja um ace e o dealer também garante no mínimo o empate, porém, é mais provável vir um 10 do que um ace. Por isso há esse drop off nos extremos da última linha, mas ele é maior no caso de a carta virada para cima ser um ace. 

iii) Para somas mais baixas, o ace poder ser 1 ou 11 ajuda mais o jogador, do que se o ace fosse 1. Isto acontece porque a flexibilização do ace faz com que o seu valor seja o melhor possível de acordo com as cartas disponíveis para o jogador. Como o valor pode chegar em 11 que é um número alto, para somas baixas ele pode fazer a diferença.
