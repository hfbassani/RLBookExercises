#### *Exercise 5.2*

#### Suppose every-visit MC was used instead of first-visit MC on the blackjack task. Would you expect the results to be very di↵erent? Why or why not?

---
#### Resposta 1:
Os resultados são os mesmos, uma vez que não é possível passar pelo mesmo estado mais de uma vez, independentemente de ser first-visit ou every-visit, o retorno é o mesmo. Para que haja diferença seria necessário que o estado fosse visitado mais de uma vez, assim adotar o every-visit afetaria o valor do estado.
