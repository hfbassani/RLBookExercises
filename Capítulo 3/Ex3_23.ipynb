**Exercício 3.23**
Give the Bellman equation for $q_*$ for the recycling robot.

A equação de otimalidade de Bellman para $q_*$ é dada por:

$
(eq. 1): q_*(s,a) = \sum_{s',r}r*p(s',r|s,a)\big[r + \gamma\max\limits_{a'}q_*(s',a')\big]
$

Para o problema do robô reciclador temos:

$S = \{h,l\}$

$A = \{s,r,re\}$

bot_mdp.PNG

Para cada combinação de $s \in S$ e $a \in A$, temos:

\\
$
q_*(h,s) = p(h|h,s)\big[r(h,s,h) + \gamma max\{q_*(h,s), q_*(h,w))\} \big] + p(l|h,s)\big[r(h,s,l) + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} \big] 
$

$
= \alpha\big[r_{search} + \gamma max\{q_*(h,s), q_*(h,w)\} \big] + (1-\alpha)\big[r_{search} + \gamma max\{q_*(h,s), q_*(h,w)\} \big]
$

\\
$
q_*(h,w) = p(h|h,w)\big[r(h,w,h) + \gamma max\{q_*(h,s), q_*(h,w)\} \big] + p(l|h,w)\big[r(h,w,l) + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} \big]
$

$
= r_{wait} + \gamma max\{q_*(h,s), q_*(h,w)\} + 0
$

\\
$
q_*(l,s) = p(l|l,s)\big[r(l,s,l) + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} \big] + p(h|l,s)\big[r(l,s,l) + \gamma max\{q_*(h,s), q_*(h,w)\} \big]
$

$
= \beta\big[r_{search} + \gamma max\{q_*(h,s), q_*(h,w)\} \big] +  (1 - \beta)\big[-3 + \gamma max\{q_*(h,s), q_*(h,w)\} \big]
$

\\
$
q_*(l,w) = p(l|l,w)\big[r(l,w,l) + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} \big] + p(h|l,w)\big[r(l,w,h) + \gamma max\{q_*(h,s), q_*(h,w)\} \big]
$

$
= r_{wait} + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} + 0
$

\\
$
q_*(l,re) = p(l|l,re)\big[r(l,re,l) + \gamma max\{q_*(l,s), q_*(l,w), q_*(l,re)\} \big] + p(h|l,re)\big[r(l,re,h) + \gamma max\{q_*(h,s), q_*(h,w)\} \big]
$

$
= 0 + \gamma max\{q_*(h,s), q_*(h,w)\}
$
